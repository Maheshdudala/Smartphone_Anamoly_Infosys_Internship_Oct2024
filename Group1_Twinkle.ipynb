```python
data = pd.read_csv('cleaned_dataset.csv')

# Function to detect outliers using IQR method
def detect_outliers(df, columns):
    outlier_indices = {}
    outlier_stats = {}
    
    for column in columns:
        # Calculate Q1, Q3, and IQR
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        
        # Calculate outlier bounds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Find outlier indices
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]
        outlier_indices[column] = outliers.index
        
        # Store statistics
        outlier_stats[column] = {
            'Q1': Q1,
            'Q3': Q3,
            'IQR': IQR,
            'Lower Bound': lower_bound,
            'Upper Bound': upper_bound,
            'Outlier Count': len(outliers),
            'Outlier Percentage': (len(outliers) / len(df)) * 100
        }
    
    return outlier_indices, outlier_stats

# Sensor columns to analyze
sensor_columns = ['Acc X', 'Acc Y', 'Acc Z', 'gyro_x', 'gyro_y', 'gyro_z', 'Speed']

# Detect outliers
outlier_indices, outlier_stats = detect_outliers(data, sensor_columns)

# Create box plots to visualize outliers
plt.figure(figsize=(15, 8))
sns.boxplot(data=data[sensor_columns])
plt.xticks(rotation=45)
plt.title('Box Plot of Sensor Readings with Outliers')
plt.tight_layout()
plt.show()

# Create a summary DataFrame of outlier statistics
outlier_summary = pd.DataFrame(outlier_stats).T
print("\nOutlier Detection Summary:")
print(outlier_summary.round(3))

# Calculate percentage of rows that contain at least one outlier
all_outlier_indices = set()
for indices in outlier_indices.values():
    all_outlier_indices.update(indices)

total_outlier_percentage = (len(all_outlier_indices) / len(data)) * 100

print(f"\nTotal rows containing at least one outlier: {len(all_outlier_indices)}")
print(f"Percentage of rows with outliers: {total_outlier_percentage:.2f}%")

# Generate detail for top outliers in each sensor
print("\nExtreme Outlier Values:")
for column in sensor_columns:
    outlier_rows = data.loc[outlier_indices[column]]
    
    # Combine the largest and smallest outliers using pd.concat instead of append
    extreme_outliers = pd.concat([outlier_rows[column].nlargest(3), outlier_rows[column].nsmallest(3)])
    
    print(f"\n{column}:")
    print(extreme_outliers.sort_values(ascending=False))
```

```python
import pandas as pd
import folium

# Load the dataset
file_path = 'cleaned_dataset.csv'  # Replace with your actual dataset path
data = pd.read_csv(file_path)

# Define a function to compute IQR and flag outliers
def iqr_outliers(column):
    Q1 = column.quantile(0.25)  # 25th percentile
    Q3 = column.quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1  # Interquartile Range

    # Calculate the bounds for detecting outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Return a boolean series where True indicates an outlier
    outliers = (column < lower_bound) | (column > upper_bound)
    return outliers

# Apply the IQR outlier detection to each relevant sensor column
for column in ['Acc X', 'Acc Y', 'Acc Z', 'gyro_x', 'gyro_y', 'gyro_z', 'Speed']:
    data[f'{column}_outlier'] = iqr_outliers(data[column])

# Filter the rows where any of the outliers columns have True
outliers_data = data[(data['Acc X_outlier']) | (data['Acc Y_outlier']) | 
                     (data['Acc Z_outlier']) | (data['gyro_x_outlier']) | 
                     (data['gyro_y_outlier']) | (data['gyro_z_outlier']) | 
                     (data['Speed_outlier'])]

# Check if the dataset contains Latitude and Longitude columns
if 'Latitude' not in data.columns or 'Longitude' not in data.columns:
    print("Latitude and Longitude columns not found in the dataset.")
else:
    # Initialize a folium map at the average location of the data points
    map_center = [outliers_data['Latitude'].mean(), outliers_data['Longitude'].mean()]
    outliers_map = folium.Map(location=map_center, zoom_start=12)

    # Add outliers as markers on the map
    for index, row in outliers_data.iterrows():
        folium.Marker(
            location=[row['Latitude'], row['Longitude']],
            popup=f"Outlier in: {', '.join([col for col in ['Acc X', 'Acc Y', 'Acc Z', 'gyro_x', 'gyro_y', 'gyro_z', 'Speed'] if row[f'{col}_outlier']])}",
            icon=folium.Icon(color='red', icon='exclamation-sign')
        ).add_to(outliers_map)

    #display it
    outliers_map
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('cleaned_dataset.csv')  # Replace with actual dataset path
```

```python
# Set up hyperparameter values to test
contamination_rates = [0.01, 0.05, 0.1]
n_estimators_values = [50, 100, 200]
```

```python
# Dictionary to store the results
results = {
    'contamination_rate': [],
    'n_estimators': [],
    'anomalies_detected': [],
    'anomaly_scores': []
}

# Function to train Isolation Forest and record results
def train_isolation_forest(data, contamination_rate, n_estimators):
    # Initialize the model with given parameters
    model = IsolationForest(contamination=contamination_rate, n_estimators=n_estimators, random_state=42)
    model.fit(data)

    # Predict anomaly scores and whether each point is an anomaly
    anomaly_scores = model.decision_function(data)
    predictions = model.predict(data)

    # Record the number of anomalies (label -1 indicates anomalies)
    anomalies_detected = np.sum(predictions == -1)

    # Store results
    results['contamination_rate'].append(contamination_rate)
    results['n_estimators'].append(n_estimators)
    results['anomalies_detected'].append(anomalies_detected)
    results['anomaly_scores'].append(anomaly_scores)

# Sensor columns to use for anomaly detection
sensor_columns = ['Acc X', 'Acc Y', 'Acc Z', 'gyro_x', 'gyro_y', 'gyro_z', 'Speed']

# Standardize the dataset for consistent scoring
data_scaled = (data[sensor_columns] - data[sensor_columns].mean()) / data[sensor_columns].std()
```

```python
# Loop over each combination of contamination rate and number of estimators
for contamination in contamination_rates:
    for n_estimators in n_estimators_values:
        train_isolation_forest(data_scaled, contamination, n_estimators)
```

```python
# Convert results to DataFrame for easy analysis
results_df = pd.DataFrame(results)

# Display summary
print("\nAnomaly Detection Results:")
print(results_df[['contamination_rate', 'n_estimators', 'anomalies_detected']])
```

```python
# Plot anomaly scores for each parameter set
plt.figure(figsize=(12, 8))

for idx, row in results_df.iterrows():
    plt.plot(row['anomaly_scores'], label=f"Cont: {row['contamination_rate']}, Est: {row['n_estimators']}")

plt.title("Anomaly Scores for Different Hyperparameter Sets")
plt.xlabel("Data Point Index")
plt.ylabel("Anomaly Score")
plt.legend(loc="upper right")
plt.tight_layout()
plt.show()
```

```python
# Print a summary for each parameter set
print("\nAnalysis Summary:")
for idx, row in results_df.iterrows():
    print(f"\nContamination Rate: {row['contamination_rate']}, Estimators: {row['n_estimators']}")
    print(f" - Anomalies Detected: {row['anomalies_detected']}")
    print(f" - Average Anomaly Score: {np.mean(row['anomaly_scores']):.3f}")
    print(f" - Minimum Anomaly Score: {np.min(row['anomaly_scores']):.3f}")
    print(f" - Maximum Anomaly Score: {np.max(row['anomaly_scores']):.3f}")
```
