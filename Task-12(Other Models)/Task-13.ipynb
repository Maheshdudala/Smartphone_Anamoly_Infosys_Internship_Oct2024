```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'Featured_dataset.csv'
data = pd.read_csv(file_path)

# Define feature columns and target (assuming target column 'label' with 0 = normal, 1 = anomaly)
feature_columns = ['Acc X', 'gyro_x', 'gyro_y', 'gyro_z', 'Acc Y', 'Acc Z']
target_column = 'label'  # Ensure your dataset has this column
X = data[feature_columns].values
y = data[target_column].values

# Define parameters for cross-validation
fold_sizes = [5, 10, 15]  # Different fold sizes to test
regularization_values = [0.1, 1, 10]  # Different C values to test for logistic regression

# Initialize results storage
results = {
    'Model': [],
    'Fold Size': [],
    'Fold': [],
    'C-Value': [],
    'Accuracy': [],
    'Precision': [],
    'Recall': []
}

# Function to evaluate Logistic Regression using K-Fold Cross-Validation
def cross_validate_logistic_regression(c_value, fold_size):
    kf = KFold(n_splits=fold_size, shuffle=True, random_state=42)
    fold_num = 1
    model = LogisticRegression(C=c_value, solver='liblinear')

    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Fit the model on the training set
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, zero_division=0)
        recall = recall_score(y_test, y_pred, zero_division=0)

        # Store results
        results['Model'].append('Logistic Regression')
        results['Fold Size'].append(fold_size)
        results['Fold'].append(fold_num)
        results['C-Value'].append(c_value)
        results['Accuracy'].append(accuracy)
        results['Precision'].append(precision)
        results['Recall'].append(recall)

        # Move to next fold
        fold_num += 1

# Run cross-validation on Logistic Regression with varying fold sizes and regularization values
for c_value in regularization_values:
    for fold_size in fold_sizes:
        cross_validate_logistic_regression(c_value, fold_size)

# Convert results to DataFrame for easier analysis
results_df = pd.DataFrame(results)

# Display results summary
print("Results Summary:")
print(results_df.groupby(['C-Value', 'Fold Size'])[['Accuracy', 'Precision', 'Recall']].mean().reset_index())

# Plot mean accuracy, precision, and recall
plt.figure(figsize=(14, 6))

# Plot Accuracy
plt.subplot(1, 3, 1)
for c_value in results_df['C-Value'].unique():
    subset = results_df[results_df['C-Value'] == c_value]
    plt.plot(subset['Fold Size'], subset['Accuracy'], marker='o', label=f"C={c_value}")
plt.xlabel("Fold Size")
plt.ylabel("Accuracy")
plt.legend(title="C-Value")
plt.title("Accuracy by Fold Size and C-Value (Logistic Regression)")

# Plot Precision
plt.subplot(1, 3, 2)
for c_value in results_df['C-Value'].unique():
    subset = results_df[results_df['C-Value'] == c_value]
    plt.plot(subset['Fold Size'], subset['Precision'], marker='o', label=f"C={c_value}")
plt.xlabel("Fold Size")
plt.ylabel("Precision")
plt.legend(title="C-Value")
plt.title("Precision by Fold Size and C-Value (Logistic Regression)")

# Plot Recall
plt.subplot(1, 3, 3)
for c_value in results_df['C-Value'].unique():
    subset = results_df[results_df['C-Value'] == c_value]
    plt.plot(subset['Fold Size'], subset['Recall'], marker='o', label=f"C={c_value}")
plt.xlabel("Fold Size")
plt.ylabel("Recall")
plt.legend(title="C-Value")
plt.title("Recall by Fold Size and C-Value (Logistic Regression)")

plt.tight_layout()
plt.show()

# Additional Plots for Detailed Analysis
plt.figure(figsize=(14, 8))

# Distribution of Accuracy by C-Value
plt.subplot(2, 2, 1)
sns.boxplot(data=results_df, x='C-Value', y='Accuracy')
plt.xlabel("C-Value")
plt.ylabel("Accuracy")
plt.title("Distribution of Accuracy by C-Value (Logistic Regression)")

# Distribution of Precision by C-Value
plt.subplot(2, 2, 2)
sns.boxplot(data=results_df, x='C-Value', y='Precision')
plt.xlabel("C-Value")
plt.ylabel("Precision")
plt.title("Distribution of Precision by C-Value (Logistic Regression)")

# Distribution of Recall by C-Value
plt.subplot(2, 2, 3)
sns.boxplot(data=results_df, x='C-Value', y='Recall')
plt.xlabel("C-Value")
plt.ylabel("Recall")
plt.title("Distribution of Recall by C-Value (Logistic Regression)")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'Featured_dataset.csv'
data = pd.read_csv(file_path)

# Define feature columns and target (assuming target column 'label' with 0 = normal, 1 = anomaly)
feature_columns = ['Acc X', 'gyro_x', 'gyro_y', 'gyro_z', 'Acc Y', 'Acc Z']
target_column = 'label'  # Ensure your dataset has this column
X = data[feature_columns].values
y = data[target_column].values

# Define parameters for cross-validation
fold_sizes = [5, 10, 15]  # Different fold sizes to test
regularization_values = [0.1, 1, 10]  # Different C values to test for logistic regression

# Initialize results storage
results = {
    'Model': [],
    'Fold Size': [],
    'Fold': [],
    'C-Value': [],
    'Accuracy': [],
    'Precision': [],
    'Recall': []
}

# Function to evaluate Logistic Regression using K-Fold Cross-Validation
def cross_validate_logistic_regression(c_value, fold_size):
    kf = KFold(n_splits=fold_size, shuffle=True, random_state=42)
    fold_num = 1
    model = LogisticRegression(C=c_value, penalty='l2', solver='liblinear')  # Use 'l2' penalty

    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Fit the model on the training set
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, zero_division=0)
        recall = recall_score(y_test, y_pred, zero_division=0)

        # Store results
        results['Model'].append('Logistic Regression')
        results['Fold Size'].append(fold_size)
        results['Fold'].append(fold_num)
        results['C-Value'].append(c_value)
        results['Accuracy'].append(accuracy)
        results['Precision'].append(precision)
        results['Recall'].append(recall)

        # Move to next fold
        fold_num += 1

# Run cross-validation on Logistic Regression with varying fold sizes and regularization values
for c_value in regularization_values:
    for fold_size in fold_sizes:
        cross_validate_logistic_regression(c_value, fold_size)

# Convert results to DataFrame for easier analysis
results_df = pd.DataFrame(results)

# Display results summary
print("Results Summary:")
print(results_df.groupby(['C-Value', 'Fold Size'])[['Accuracy', 'Precision', 'Recall']].mean().reset_index())

# Plot mean accuracy, precision, and recall
plt.figure(figsize=(14, 6))

# Plot Accuracy
plt.subplot(1, 3, 1)
for c_value in results_df['C-Value'].unique():
    subset = results_df[results_df['C-Value'] == c_value]
    plt.plot(subset['Fold Size'], subset['Accuracy'], marker='o', label=f"C={c_value}")
plt.xlabel("Fold Size")
plt.ylabel("Accuracy")
plt.legend(title="C-Value")
plt.title("Accuracy by Fold Size and C-Value (Logistic Regression)")

# Plot Precision
plt.subplot(1, 3, 2)
for c_value in results_df['C-Value'].unique():
    subset = results_df[results_df['C-Value'] == c_value]
    plt.plot(subset['Fold Size'], subset['Precision'], marker='o', label=f"C={c_value}")
plt.xlabel("Fold Size")
plt.ylabel("Precision")
plt.legend(title="C-Value")
plt.title("Precision by Fold Size and C-Value (Logistic Regression)")

# Plot Recall
plt.subplot(1, 3, 3)
for c_value in results_df['C-Value'].unique():
    subset = results_df[results_df['C-Value'] == c_value]
    plt.plot(subset['Fold Size'], subset['Recall'], marker='o', label=f"C={c_value}")
plt.xlabel("Fold Size")
plt.ylabel("Recall")
plt.legend(title="C-Value")
plt.title("Recall by Fold Size and C-Value (Logistic Regression)")

plt.tight_layout()
plt.show()

# Additional Plots for Detailed Analysis
plt.figure(figsize=(14, 8))

# Distribution of Accuracy by C-Value
plt.subplot(2, 2, 1)
sns.boxplot(data=results_df, x='C-Value', y='Accuracy')
plt.xlabel("C-Value")
plt.ylabel("Accuracy")
plt.title("Distribution of Accuracy by C-Value (Logistic Regression)")

# Distribution of Precision by C-Value
plt.subplot(2, 2, 2)
sns.boxplot(data=results_df, x='C-Value', y='Precision')
plt.xlabel("C-Value")
plt.ylabel("Precision")
plt.title("Distribution of Precision by C-Value (Logistic Regression)")

# Distribution of Recall by C-Value
plt.subplot(2, 2, 3)
sns.boxplot(data=results_df, x='C-Value', y='Recall')
plt.xlabel("C-Value")
plt.ylabel("Recall")
plt.title("Distribution of Recall by C-Value (Logistic Regression)")

plt.tight_layout()
plt.show()
```
